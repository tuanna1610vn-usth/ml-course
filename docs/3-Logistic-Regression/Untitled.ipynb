```python
from sklearn.metrics import accuracy_score, confusion_matrix
import pandas as pd

# Create a small dataset
data = [
    {"email": "Win a free iPhone now!", "true": "spam", "pred": "spam"},
    {"email": "Meeting at 10am", "true": "not spam", "pred": "not spam"},
    {"email": "Limited offer just for you", "true": "spam", "pred": "not spam"},  # mistake
    {"email": "Project update attached", "true": "not spam", "pred": "not spam"},
    {"email": "Congratulations, you won!", "true": "spam", "pred": "spam"},
    {"email": "Lunch tomorrow?", "true": "not spam", "pred": "spam"},  # mistake
    {"email": "Invoice for last month", "true": "not spam", "pred": "not spam"},
]

df = pd.DataFrame(data)
df
```

```python
acc = accuracy_score(df["true"], df["pred"])
print("Accuracy:", acc)
```

```python
cm = confusion_matrix(df["true"], df["pred"], labels=["spam", "not spam"])
print(cm)
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(df["true"], df["pred"], labels=["spam", "not spam"])

cm_df = pd.DataFrame(
    cm,
    index=["Actual Spam", "Actual Not Spam"],
    columns=["Pred Spam", "Pred Not Spam"]
)

print(cm_df)
```

```python
import matplotlib.pyplot as plt

plt.imshow(cm)
plt.title("Confusion Matrix")

plt.xticks([0,1], ["Spam", "Not Spam"])
plt.yticks([0,1], ["Spam", "Not Spam"])

plt.xlabel("Predicted")
plt.ylabel("Actual")

for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j], ha="center", va="center", color="white")

plt.show()
```
