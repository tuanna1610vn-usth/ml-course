# Exercise 1 — Linear Regression


## Part A: Understanding LR

Imagine we have data from a small restaurant.

For each table, we know:

- the total bill
- the tip left by the customer

We want to predict the tip from the bill using a straight line:

`predicted tip = weight × bill + bias`

Your task is to change `weight` and `bias` until the line fits the data as well as possible.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample restaurant data
# x = total bill in dollars
# y = tip in dollars

x = np.array([10, 12, 15, 18, 20, 22, 25, 28, 30, 35, 40, 45])

y = np.array([1.8, 2.0, 2.7, 3.0, 3.6, 3.5, 4.2, 4.8, 5.0, 5.8, 6.5, 7.0])
```

```python
# You should change only these two numbers
weight = -0.1
bias = 3.0
```

```python
# Predictions
y_pred = weight * x + bias

# Residuals and losses
residuals = y - y_pred

l1 = np.sum(np.abs(residuals))
l2 = np.sum(residuals ** 2)
mse = np.mean(residuals ** 2)
rmse = np.sqrt(mse)

# Plot
plt.figure(figsize=(10, 6))

plt.scatter(x, y, s=90, label="Real data")

# Regression line
x_line = np.linspace(0, 50, 100)
y_line = weight * x_line + bias

plt.plot(
    x_line,
    y_line,
    linewidth=3,
    label=f"Model: tip = {weight:.2f} × bill + {bias:.2f}"
)

# Residual lines
for xi, yi, ypi in zip(x, y, y_pred):
    plt.plot([xi, xi], [yi, ypi], linewidth=2, alpha=0.7)

plt.title("Linear Regression: Predicting the Tip")
plt.xlabel("Total bill, x ($)")
plt.ylabel("Tip, y ($)")

plt.xlim(0, 50)
plt.ylim(0, 9)
plt.grid(True, alpha=0.3)
plt.legend()

loss_text = (
    f"L1 loss:  {l1:.2f}\n"
    f"L2 loss:  {l2:.2f}\n"
    f"MSE:      {mse:.2f}\n"
    f"RMSE:     {rmse:.2f}"
)

plt.text(
    0.05,
    0.95,
    loss_text,
    transform=plt.gca().transAxes,
    verticalalignment="top",
    bbox=dict(boxstyle="round", alpha=0.15),
    fontsize=12
)

plt.show()
```

```python
# Compute the best linear regression solution automatically

best_weight, best_bias = np.polyfit(x, y, 1)

y_best = best_weight * x + best_bias
best_mse = np.mean((y - y_best) ** 2)
best_rmse = np.sqrt(best_mse)

print(f"Best weight: {best_weight:.3f}")
print(f"Best bias:   {best_bias:.3f}")
print(f"Best MSE:    {best_mse:.3f}")
print(f"Best RMSE:   {best_rmse:.3f}")
```

# Exercise 2 — ML Key Challenges


In this exercise, you will train several models and observe three cases:

1. Underfitting — model too simple
2. Good fit — model captures the pattern
3. Overfitting — model memorizes the noise

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(42)

# Input values
X = np.linspace(0, 10, 80).reshape(-1, 1)

# True pattern
y_true = np.sin(X).ravel()

# Add noise
noise = np.random.normal(0, 0.25, size=y_true.shape)
y = y_true + noise

plt.figure(figsize=(8, 5))
plt.scatter(X, y, label="Noisy data")
plt.plot(X, y_true, linewidth=2, label="True pattern")
plt.title("Noisy Dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("Training examples:", len(X_train))
print("Test examples:", len(X_test))
```

```python
degrees = [1, 4, 15]

results = []

x_plot = np.linspace(0, 10, 300).reshape(-1, 1)

plt.figure(figsize=(15, 4))

for i, degree in enumerate(degrees, 1):
    model = make_pipeline(
        PolynomialFeatures(degree),
        LinearRegression()
    )
    
    model.fit(X_train, y_train)
    
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    y_plot = model.predict(x_plot)
    
    train_mse = mean_squared_error(y_train, y_train_pred)
    test_mse = mean_squared_error(y_test, y_test_pred)
    
    results.append({
        "degree": degree,
        "train_mse": train_mse,
        "test_mse": test_mse
    })
    
    plt.subplot(1, 3, i)
    plt.scatter(X_train, y_train, label="Training data", alpha=0.7)
    plt.scatter(X_test, y_test, label="Test data", alpha=0.7)
    plt.plot(x_plot, y_plot, linewidth=3, label=f"Degree {degree}")
    plt.title(f"Polynomial Degree {degree}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

results_df = pd.DataFrame(results)
results_df
```

```python
plt.figure(figsize=(8, 5))

plt.plot(results_df["degree"], results_df["train_mse"], marker="o", label="Training MSE")
plt.plot(results_df["degree"], results_df["test_mse"], marker="o", label="Test MSE")

plt.title("Training Error vs Test Error")
plt.xlabel("Model complexity: polynomial degree")
plt.ylabel("MSE")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
